```python
import numpy as np

def Train_Test_Splite(data , test_ratio):
    np.random.seed(42)
    shuffeled = np.random.permutation(len(data))
    test_size = int(len(data)*test_ratio)
    train_set = shuffeled[test_size:]
    test_set = shuffeled[:test_size]

    return data.iloc[train_set] , data.iloc[test_set]
train_set , test_set = Train_Test_Splite(house , 0.2)
```

```python
house = train_set.drop("MEDV" , axis=1)
house_lables = train_set["MEDV"].copy()
```

```python
house.shape
```

```python
from sklearn.preprocessing import StandardScaler 
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

my_pipeline = Pipeline (
    [("Imputer", SimpleImputer(strategy= "median") ),
    ("StandarScaler" ,StandardScaler())
    ]
       
)
```

```python
house_new  = my_pipeline.fit_transform(house)
```

```python
from sklearn.ensemble import RandomForestRegressor 
model = RandomForestRegressor()
```

```python
model.fit(house_new, house_lables)
```

```python
some_data = house.iloc[:5]
some_labels = house_lables.iloc[:5]
```

```python
prepared_data = my_pipeline.fit_transform(some_data)
```

```python
model.predict(prepared_data)
```

```python
list(some_labels)
```

```python
from sklearn.metrics import mean_squared_error
error_detection = model.predict(house_new)
linear_mse  = mean_squared_error(house_lables, error_detection)
linear_rmse = np.sqrt(linear_mse)
```

```python
linear_rmse
```

## Cross Validation 

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(model , house_new ,house_lables, scoring = "neg_mean_squared_error", cv =20)
rmse = np.sqrt(-score)
```

```python
rmse
```

```python
def prediction_model(value) :
    print("score:" , value )
    print("mean: " , value.mean())
    print("standard_deviation :" , value.std())
```

```python
prediction_model(rmse)
```

```python
x = test_set.drop("MEDV",axis=1)
y = test_set["MEDV"].copy()
mydata = my_pipeline.fit_transform(x)
prediction = model.predict(mydata)
mse = mean_squared_error(y,prediction)
rmse = np.sqrt(mse)
```
